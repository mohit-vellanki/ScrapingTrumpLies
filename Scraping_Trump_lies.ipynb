```python
import requests  
req = requests.get('https://www.nytimes.com/interactive/2017/06/23/opinion/trumps-lies.html')
```

```python
print(req.url)
print(req.text[0:100])
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
#searching for the attributes
results = soup.find_all('span', attrs={'class':'short-desc'})  
len(results)
```

```python
first_result = results[0]
```

```python
print(first_result)
```

```python
first_result.find('strong').text[0:-1]+ ', 2017'
```

```python
first_result.contents[1][1:-2]
first_result.find('a').text[1:-1]
```

```python
records = []  
for result in results:  
    date = result.find('strong').text[0:-1] + ', 2017'
    lie = result.contents[1][1:-2]
    explanation = result.find('a').text[1:-1]
    url = result.find('a')['href']
    records.append((date, lie, explanation, url))
```

```python
records[1]
```

```python
import pandas as pd
df = pd.DataFrame(records, columns = ['date', 'lie', 'truth', 'url'])
```

```python
df.head(4)
```

```python
df.to_csv("Trump_lies_truth.csv", index = False)
```
